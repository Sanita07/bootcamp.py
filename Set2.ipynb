```python
sentence = input("Enter a sentence: ")
words = sentence.split()
freq = {}
for word in words:
    if word in freq:
        freq[word] = freq[word] + 1
    else:
        freq[word] = 1
print("Word Frequencies:")
for i in freq:
    print(i, ":", freq[i])
```

```python
d1 = {'a': 10, 'b': 20, 'c': 30}
d2 = {'b': 15, 'c': 25, 'd': 40}
result = {}
for key in d1:
    result[key] = d1[key]
for key in d2:
    if key in result:
        result[key] = result[key] + d2[key]
    else:
        result[key] = d2[key]
print("Merged Dictionary:")
for key in result:
    print(key, ":", result[key])
```
